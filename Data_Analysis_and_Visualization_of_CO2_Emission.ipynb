**Average Carbon Emission of the respective countries in the year '2000' in descending order**

```python
dataframe.groupby('Country Name')['2000'].mean().sort_values(ascending=False)
```

**Plotting the Graph for the same using Seaborn and Matplotlib**

```python
sb.barplot(x='Country Name',y='2000',data=dataframe)
plt.title('Carbon Emission of the respective countries (Year = 2000)')
plt.show()
```

**Average carbon Emission in 1997**

```python
dataframe.groupby('Country Name')['1997'].mean()
```

```python
sb.barplot(x='Country Name',y='1997',data=dataframe)
plt.title('Carbon Emission ')
plt.show()
```
